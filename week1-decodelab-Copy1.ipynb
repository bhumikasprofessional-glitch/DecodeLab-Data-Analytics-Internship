```python
import pandas as pd

# Bilkul simple tarika, koi lamba path nahi chahiye!
df = pd.read_csv("retail_sales_dataset.csv")

# Data check karne ke liye
df.head()
```

```python

df.info()
```

```python
# Step 1: Checking for Missing Values
print("Missing Values:")
print(df.isnull().sum())
# Note: No missing values were found in this dataset.
```

```python
# Check karein kitne duplicate rows hain
duplicates_count = df.duplicated().sum()
print("Total Duplicate Rows:", duplicates_count)

# Agar koi duplicate row milti hai toh use remove karein
df.drop_duplicates(inplace=True)
print("Duplicates handle karne ke baad data ka shape:", df.shape)
```

```python
# Date column ka format sahi (datetime format) karein
df['Date'] = pd.to_datetime(df['Date'])

# Check karne ke liye ki format sahi hua ya nahi
print("Date column ka naya data type:", df['Date'].dtype)
```

```python
# Text columns ko clean aur capitalize karein
df['Gender'] = df['Gender'].str.strip().str.capitalize()
df['Product Category'] = df['Product Category'].str.strip().str.title()

# Final clean data ka ek baar structure dekhne ke liye
df.info()
```

```python
# Saaf data ko nayi CSV file mein save karein
df.to_csv("Cleaned_Retail_Sales_Dataset.csv", index=False)
print("Mubarak ho! Week 1 ka Project 1 successfully ready ho gaya hai.")
```
